```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../Dataset/processed/zomato_featured.csv")
```

# Business Insights

## Objective

The objective of this notebook is to convert analytical findings into actionable business recommendations.

The analysis focuses on identifying customer behavior, restaurant performance, pricing strategies, service preferences, and market opportunities that can help Zomato improve customer satisfaction and restaurant recommendations.

## Business Insight 1

### Business Question
Do restaurants offering online ordering receive higher customer ratings?

```python
online_rating = df.groupby("online_order")["rate"].mean().round(2)
online_rating
```

### Observation

Restaurants offering online ordering generally achieve higher average ratings.

### Recommendation

Encourage restaurant partners to enable online ordering to improve customer convenience and increase customer satisfaction.

## Business Insight 2

### Business Question
Does table booking improve customer satisfaction?

```python
booking_rating = df.groupby("book_table")["rate"].mean().round(2)
booking_rating
```

### Observation

Restaurants providing table booking services generally receive higher ratings.

### Recommendation

Restaurants targeting premium customers should enable table booking to improve the dining experience.

## Business Insight 3

### Business Question
Which locations have the highest restaurant concentration?

```python
df["location"].value_counts().head(10)
```

### Observation

A few locations contain a significantly higher number of restaurants, indicating intense market competition.

### Recommendation

Restaurants entering these markets should differentiate themselves through unique cuisines, pricing, or customer experience.

## Business Insight 4

### Business Question
Which cuisines dominate the Bangalore restaurant market?

```python
df["cuisines"].value_counts().head(10)
```

### Observation

North Indian, Chinese, and South Indian cuisines are among the most frequently offered.

### Recommendation

Marketing campaigns should prioritize popular cuisines while also promoting niche cuisine categories to attract diverse customer segments.

## Business Insight 5

### Business Question
How does pricing influence restaurant ratings?

```python
df.groupby("Price_Category")["rate"].mean().round(2)
```

### Observation

Premium-priced restaurants do not always achieve the highest customer ratings.

### Recommendation

Restaurants should focus on delivering consistent service quality instead of relying solely on premium pricing.

## Business Insight 6

### Business Question
Which restaurant types receive the highest customer ratings?

```python
df.groupby("rest_type")["rate"].mean().sort_values(ascending=False).head(10)
```

### Observation

Certain restaurant formats consistently outperform others in customer satisfaction.

### Recommendation

Restaurant partners can benchmark these successful formats to improve service offerings.

## Business Insight 7

### Business Question
Which restaurants generate the highest customer engagement?

```python
df.nlargest(10, "votes")[["name", "votes", "rate"]]
```

### Observation

A small number of restaurants account for a large share of total customer engagement.

### Recommendation

Highly engaged restaurants should receive greater visibility through featured listings and promotional campaigns.

## Business Insight 8

### Business Question
Which price category dominates the restaurant market?

```python
df["Price_Category"].value_counts()
```

### Observation

Budget and Mid-Range restaurants make up the majority of the market.

### Recommendation

Discount campaigns and loyalty programs should primarily target these customer segments.

## Business Insight 9

### Business Question
Which restaurants have high ratings but relatively low customer visibility?

```python
hidden_gems = df[
    (df["rate"] >= 4.5) &
    (df["votes"] < 100)
][["name", "location", "rate", "votes"]]

hidden_gems.head(10)
```

### Observation

Several highly rated restaurants have received relatively few customer votes.

### Recommendation

Zomato can improve restaurant discovery by recommending these "hidden gems" to users.

## Business Insight 10

### Business Question
Which locations are suitable for future restaurant expansion?

```python
expansion = (
    df.groupby("location")
      .agg(
          Restaurants=("name", "count"),
          Average_Rating=("rate", "mean")
      )
      .sort_values("Average_Rating", ascending=False)
)

expansion.head(10)
```

### Observation

Some locations maintain high average ratings while having relatively fewer restaurants, indicating potential market opportunities.

### Recommendation

Restaurant expansion strategies should prioritize these locations to capture growing customer demand with lower competition.

# Final Business Recommendations

Based on the analysis, the following recommendations are proposed:

- Expand online ordering adoption across restaurant partners.
- Promote highly rated restaurants with low customer visibility.
- Focus restaurant expansion in high-rated, low-competition locations.
- Encourage premium restaurants to improve service quality alongside pricing.
- Increase marketing for high-performing cuisines and restaurant types.
- Recommend hidden gems to improve customer discovery.
- Use customer ratings and engagement metrics to personalize restaurant recommendations.
- Continue monitoring customer preferences through ratings, votes, and service offerings.
