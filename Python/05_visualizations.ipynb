```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../Dataset/processed/zomato_featured.csv")
```

```python
top_locations = (
    df["location"]
    .value_counts()
    .head(10)
)

plt.figure(figsize=(10,6))
top_locations.plot(kind="bar")
plt.title("Top 10 Restaurant Locations")
plt.xlabel("Location")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["rate"].dropna(), bins=15)
plt.title("Distribution of Restaurant Ratings")
plt.xlabel("Rating")
plt.ylabel("Number of Restaurants")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,5))
df["Price_Category"].value_counts().plot(kind="bar")
plt.title("Restaurant Price Categories")
plt.xlabel("Price Category")
plt.ylabel("Number of Restaurants")
plt.tight_layout()
plt.show()
```

```python
online_rating = (
    df.groupby("online_order")["rate"]
    .mean()
)

plt.figure(figsize=(6,5))
online_rating.plot(kind="bar")
plt.title("Average Rating by Online Ordering")
plt.xlabel("Online Order")
plt.ylabel("Average Rating")
plt.tight_layout()
plt.show()
```

```python
booking_rating = (
    df.groupby("book_table")["rate"]
    .mean()
)

plt.figure(figsize=(6,5))
booking_rating.plot(kind="bar")
plt.title("Average Rating by Table Booking")
plt.xlabel("Book Table")
plt.ylabel("Average Rating")
plt.tight_layout()
plt.show()
```

```python
top_cuisines = (
    df["cuisines"]
    .value_counts()
    .head(10)
)

plt.figure(figsize=(12,6))
top_cuisines.plot(kind="bar")
plt.title("Top 10 Cuisines")
plt.xlabel("Cuisine")
plt.ylabel("Restaurants")
plt.xticks(rotation=60)
plt.tight_layout()
plt.show()
```

```python
top_types = (
    df["rest_type"]
    .value_counts()
    .head(10)
)

plt.figure(figsize=(10,6))
top_types.plot(kind="bar")
plt.title("Top Restaurant Types")
plt.xlabel("Restaurant Type")
plt.ylabel("Restaurants")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["votes"], bins=30)
plt.title("Distribution of Customer Votes")
plt.xlabel("Votes")
plt.ylabel("Restaurants")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["approx_cost"].dropna(), bins=25)
plt.title("Distribution of Approximate Cost")
plt.xlabel("Cost")
plt.ylabel("Restaurants")
plt.tight_layout()
plt.show()
```

```python
top_restaurants = (
    df.nlargest(10, "votes")
)

plt.figure(figsize=(10,6))
plt.bar(top_restaurants["name"], top_restaurants["votes"])
plt.title("Top 10 Restaurants by Votes")
plt.xlabel("Restaurant")
plt.ylabel("Votes")
plt.xticks(rotation=60)
plt.tight_layout()
plt.show()
```

```python
corr = df[["rate","votes","approx_cost","Popularity_Score"]].corr()

plt.figure(figsize=(6,5))
plt.imshow(corr, interpolation="nearest")
plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("Correlation Matrix")
plt.tight_layout()
plt.show()
```
